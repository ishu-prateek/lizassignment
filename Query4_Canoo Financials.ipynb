```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://companiesmarketcap.com/canoo/total-liabilities/")
print(r.status_code)
```

```python
soup=BeautifulSoup(r.text,'html.parser')
```

```python
table = soup.find("table", class_="table")
f = pd.DataFrame(
    columns=["Year", "Total_liabilities", "Diff_liabilities"]
)
mylist = []
for table_row in table.tbody.find_all("tr"):
    table_columns = table_row.find_all("td")

    if table_columns != []:
        year = table_columns[0].text.strip()
        liabilities = table_columns[1].text.strip()
        diff = table_columns[2].text.strip()

        mylist.append([year,liabilities,diff])

df_liability = pd.DataFrame(
    mylist,
    columns=["Year", "Total_liabilities", "Diff_liabilities"]
)
```

```python
r7=requests.get("https://companiesmarketcap.com/canoo/net-assets/")
soup7=BeautifulSoup(r7.text,'html.parser')
table = soup7.find("table", class_="table")
f = pd.DataFrame(
    columns=["Year", "Net_assets", "net_Diff"]
)
mylist = []
for table_row in table.tbody.find_all("tr"):
    table_columns = table_row.find_all("td")

    if table_columns != []:
        year = table_columns[0].text.strip()
        net_assets = table_columns[1].text.strip()
        diff = table_columns[2].text.strip()

        mylist.append([year,net_assets,diff])

df_net_assets = pd.DataFrame(
    mylist,
    columns=["Year", "Net_assets", "net_Diff"]
)
```

```python
df_net_assets
```

```python
def convert_date(obj):
    if len(obj)>5:
        return obj[:4]
    else:
        return obj
```

```python
df_net_copy=df_net_assets.copy()
```

```python
df_net_copy['Year']=df_net_copy['Year'].apply(convert_date)
```

```python
df_assets["Year"]=df_assets["Year"].apply(convert_date)
df_assets
```

```python
df_cash["Year"]=df_cash["Year"].apply(convert_date)
df_cash
```

```python
df_liability["Year"]=df_liability["Year"].apply(convert_date)
df_liability
```

```python
df_pb["Year"]=df_pb["Year"].apply(convert_date)
df_pb
```

```python
df_eps
```

```python
df_eps_pb=pd.merge(df_eps,df_pb,on="Year",how="left")
df_eps_pb_lia=pd.merge(df_eps_pb,df_liability,how="left",on="Year")
df_eps_pb_liaca=pd.merge(df_eps_pb_lia,df_cash,on="Year",how="left")
df_eps_pb_liacaps=pd.merge(df_eps_pb_liaca,df_ps,on="Year",how="left")
df_eps_pb_liacapspe=pd.merge(df_eps_pb_liacaps,df_pe,on="Year",how="left")
df_eps_pb_liacapspe_assets=pd.merge(df_eps_pb_liacapspe,df_assets,on="Year",how="left")
df_eps_pb_liacapspe_assets_debt=pd.merge(df_eps_pb_liacapspe_assets,df_debt,on="Year",how="left")
df_result=pd.merge(df_eps_pb_liacapspe_assets_debt,df_net_assets,on="Year",how="left")
```
